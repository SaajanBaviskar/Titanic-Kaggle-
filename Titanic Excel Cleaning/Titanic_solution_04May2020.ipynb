```python
import pandas as pd
```

```python
train = pd.read_csv("excel_train.csv")
train.head()
```

```python
validation = pd.read_csv("excel_validation.csv")
validation.head()
```

```python
test = pd.read_csv("excel_test.csv")
test.head()
```

```python
X_train = train.drop(['PassengerId','Survived'],axis = 1)
```

```python
y_train = train['Survived']
```

```python
X_val = train.drop(['PassengerId','Survived'],axis = 1)
```

```python
y_val = train['Survived']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logMod = LogisticRegression(solver = 'liblinear')
```

```python
logMod.fit(X_train,y_train)
```

```python
logMod.score(X_train,y_train)
```

```python
logMod.coef_
```

```python
logMod.intercept_
```

## Evaluation of Test

```python
X_val.head(1)
```

```python
y_val_pred = logMod.predict(X_val)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_val,y_val_pred))
```

```python
logMod.score(X_val,y_val)
```

## Prediction of Test

```python
X_test = test.drop(['PassengerId'],axis=1)
```

```python
y_test_pred = logMod.predict(X_test)
```

```python
AllSub = pd.DataFrame({'PassengerId':test["PassengerId"],
                      'Survived':y_test_pred})

AllSub.to_csv("Submission.csv", index = False)
```
